 1. What is the difference between a function and a method in Python?


The primary difference between a function and a method in Python lies in how they are associated with objects and how they are called:

Function:

A function is a block of code that performs a specific task and is defined using the def keyword.
It is not tied to any particular object and can be called independently.

Method:

A method is a function that is associated with an object.
It is called on an object and can access the object's attributes and other methods.
Methods are defined within a class and take the instance (self) as their first parameter.

2. Explain the concept of function arguments and parameters in Python.

In Python, arguments and parameters are essential concepts related to passing data to functions.

Parameters:

These are the placeholders or variables defined in the function header (within the parentheses) when a function is declared.
They act as inputs that the function expects to receive when it is called.

Arguments:

These are the actual values passed to the function when it is called.
They fill the placeholders (parameters) and provide the function with the necessary data.


3. What are the different ways to define and call a function in Python?



**Different ways to define functions in Python:**

- **Regular Function**  
  Defined using the `def` keyword.  
  ```python
  def greet(name):
      return f"Hello, {name}!"
  ```

- **Lambda Function**  
  A single-expression anonymous function defined using the `lambda` keyword.  
  ```python
  square = lambda x: x ** 2
  ```

- **Function with Default Arguments**  
  Parameters can have default values.  
  ```python
  def greet_default(name="Guest"):
      return f"Hello, {name}!"
  ```

- **Function with Variable-Length Arguments**  
  Use `*args` for positional arguments and `**kwargs` for keyword arguments.  
  ```python
  def describe_person(*args, **kwargs):
      print("Positional arguments:", args)
      print("Keyword arguments:", kwargs)
  ```

- **Nested Function**  
  Functions can be defined inside other functions.  
  ```python
  def outer_function():
      def inner_function():
          return "Hello from inner function!"
      return inner_function()
  ```

- **Recursive Function**  
  A function can call itself until a base condition is reached.  
  ```python
  def factorial(n):
      return 1 if n == 0 else n * factorial(n - 1)
  ```

- **Class Method**  
  Functions defined in a class operate on its instances.  
  ```python
  class Greeter:
      def greet(self, name):
          return f"Hello, {name}!"
  ```

---

**Different ways to call functions in Python:**

- **Regular Function Call**  
  Call by passing arguments.  
  ```python
  print(greet("Arbaz"))  # Output: Hello, Arbaz!
  ```

- **Lambda Function Call**  
  Call directly or through a variable.  
  ```python
  print(square(5))  # Output: 25
  ```

- **Calling Functions with Default Arguments**  
  Call without or with arguments.  
  ```python
  print(greet_default())       # Output: Hello, Guest!
  print(greet_default("Arbaz"))  # Output: Hello, Arbaz!
  ```

- **Calling Functions with Variable-Length Arguments**  
  Pass any number of arguments.  
  ```python
  describe_person(25, "Arbaz", city="Mumbai", hobby="coding")
  # Output:
  # Positional arguments: (25, 'Arbaz')
  # Keyword arguments: {'city': 'Mumbai', 'hobby': 'coding'}
  ```

- **Calling Nested Functions**  
  Call the outer function, which invokes the inner function.  
  ```python
  print(outer_function())  # Output: Hello from inner function!
  ```

- **Calling Recursive Functions**  
  Pass the base case and observe recursive behavior.  
  ```python
  print(factorial(5))  # Output: 120
  ```

- **Calling Class Methods**  
  Create an instance of the class and call its method.  
  ```python
  obj = Greeter()
  print(obj.greet("Arbaz"))  # Output: Hello, Arbaz!
  ```



 4. What is the purpose of the `return` statement in a Python function?


The return statement in a Python function is used to specify the value that the function will output when called, allowing the returned value to be stored in variables or used in further operations. It also terminates the execution of the function, ensuring any code written after the return statement is not executed. Additionally, it enables returning multiple values, typically as a tuple, which can be unpacked into separate variables. If a return statement is used without any value, or if it is omitted entirely, the function implicitly returns None. This makes the return statement a vital part of function behavior, enabling communication between a function and the rest of the program.

 6. Explain the concept of generators in Python and how they are defined.

Generators in Python are a type of iterable, like lists or tuples, but they allow you to generate values on the fly and are more memory-efficient. Unlike lists, which store all their elements in memory, generators generate one item at a time as you iterate over them, which makes them ideal for large datasets or streams of data.

A generator is defined using a function with the yield keyword. The yield statement returns a value from the generator function but does not terminate the function. The state of the function is saved, so when the generator is called again, it resumes from where it left off, continuing the iteration.

 7. What are the advantages of using generators over regular functions?


Generators offer several advantages over regular functions, especially when dealing with large data sets or streams of data. Some key advantages include:

Memory Efficiency: Generators yield one item at a time and do not store the entire sequence in memory. This makes them highly efficient for working with large datasets or infinite sequences, as they only generate values on demand.

Example: A list containing 1 million numbers would require significant memory, but a generator would only generate one number at a time, reducing memory usage.

Lazy Evaluation: Generators use lazy evaluation, meaning they generate values only when they are needed. This can be beneficial when processing large amounts of data, as you avoid the overhead of generating and storing the entire dataset at once.

Example: When processing large logs or data from a stream, you can start consuming the data immediately without waiting for the entire set to be processed.

Improved Performance: Since generators produce values one at a time, they can improve performance in cases where the entire dataset is not needed or where the computation of each value is independent. Regular functions return all values at once, which can be slower and more resource-intensive.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- **`map()`**: The `map()` function applies a given function to each item of an iterable (e.g., a list or tuple) and returns an iterator with the results. It can also take multiple iterables and apply the function to elements at corresponding positions.  
  Example: Squaring each number in a list.  
  ```python
  numbers = [1, 2, 3, 4]
  result = map(lambda x: x ** 2, numbers)  
  print(list(result))  # Output: [1, 4, 9, 16]
  ```

- **`reduce()`**: The `reduce()` function, from the `functools` module, takes a function and applies it cumulatively to the items of an iterable, reducing it to a single value. It processes the iterable two elements at a time, applying the function to the first two, then the result with the next, and so on.  
  Example: Finding the product of all numbers in a list.  
  ```python
  from functools import reduce
  numbers = [1, 2, 3, 4]
  result = reduce(lambda x, y: x * y, numbers)  
  print(result)  # Output: 24 (1 * 2 * 3 * 4)
  ```

- **`filter()`**: The `filter()` function filters elements from an iterable based on a condition defined in a function. It returns an iterator containing only those elements for which the function returns `True`.  
  Example: Selecting even numbers from a list.  
  ```python
  numbers = [1, 2, 3, 4, 5, 6]
  result = filter(lambda x: x % 2 == 0, numbers)  
  print(list(result))  # Output: [2, 4, 6]
  ```

### Summary:
- `map()` transforms items in an iterable.
- `reduce()` reduces an iterable to a single value.
- `filter()` selects items from an iterable based on a condition.

 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n == 1:
        return f"{n} is neither a prime nor composite"
    elif n <1:
      return "{} not a valid number to check",format(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return f"{n} is not a prime"
    return f"{n} is a prime"

# Test the function with numbers from 1 to 200
for number in range(1, 201):
    print(is_prime(number))
````

 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms

````python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.n <= 0:
            raise StopIteration
        current = self.a
        self.a, self.b = self.b, self.a + self.b
        self.n -= 1
        return current

fibonacci_sequence = FibonacciIterator(10)
for num in fibonacci_sequence:
    print(num)
````

7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
file_content = [
    "Hello, this is the first line.",
    "This is the second line.",
    "And this is the third line.",
    "This is the fourth line.",
]


def read_file_line_by_line(content):
    for line in content:
        yield line.strip()


for line in read_file_line_by_line(file_content):
    print(line)
````

 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

````python
celsius_temperatures = [30,45,32,25]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
````

 11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
 Write a Python program using lambda and map

````python
orders = [
    [101, 4,40.00],
    [202, 5, 56.80],
    [302, 3, 32.95],
    [402, 3, 24.99],
]
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)
result = list(map(lambda order: calculate_order(order), orders))

print(result)
````
